```python
import pandas as pd
import numpy as np
from astropy.table import Table
```

```python
# Specobj ID for gals
mpa_specobj_df = pd.read_csv('/Users/abry/data/MPA_JHU/DR4/specobj_id_vs_num.dat.gz',
                            sep='\s+', header=None, usecols=[1], names=['specobj_id'])
# Spec median SNR
mpa_spec_info = Table.read('/Users/abry/data/MPA_JHU/DR4/gal_info_dr4_v5_1b.fit.gz')
mpa_spec_sn_df = mpa_spec_info[['SN_MEDIAN']].to_pandas().rename(columns={'SN_MEDIAN': 'spec_sn'})

# Stellar Metal
mpa_z_star_df = pd.read_csv('/Users/abry/data/MPA_JHU/DR4/all_stat_z_log.dat.gz', 
                            sep='\s+', header=None, usecols=[5], names=['log_z_star'])
# mpa_z_star_df = mpa_z_star_df.replace(-99, np.nan).dropna().reset_index(drop=True)  
# Stellar Age
mpa_age_star_df = pd.read_csv(
    '/Users/abry/data/MPA_JHU/DR4/all_stat_age.dat.gz',
    header=None, sep='\s+',usecols=[5],names=['log_lw_age'],)
# mpa_age_star_df = mpa_age_star_df[mpa_age_star_df.log_lw_age>=0].reset_index(drop=True)
# Stellar Mass & Redshift
mpa_z_df = pd.read_csv('/Users/abry/data/MPA_JHU/DR4/newmasses.dat_dr4.gz',
                          sep='\s+', header=None, usecols=[0, 1, 2, 4, 8], names=['plate', 'mjd', 'fiberid','z', 'log_m_star'])


# SFR
mpa_sfr = Table.read('/Users/abry/data/MPA_JHU/DR4/sfr_dr4_v2_tot.fit.gz')
mpa_sfr_df = mpa_sfr[['MEDIAN']].to_pandas().rename(columns={'MEDIAN': 'log_sfr'})
# Index for SF galaxies
mpa_sfr_info = Table.read('/Users/abry/data/MPA_JHU/DR4/sfr_dr4_v2_info.fit.gz')
mpa_sf_idx_df = mpa_sfr_info[['PLATEID', 'MJD', 'FIBERID', 'CLASS']].to_pandas().rename(columns={'CLASS': 'sf_class'})
mpa_sfr_idx_df = pd.concat([mpa_sf_idx_df, mpa_sfr_df], axis=1)
mpa_sfr_idx_df
```
